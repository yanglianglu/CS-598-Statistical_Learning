```python
import requests
```

```python
api_key = open('data/api_key').read()
api_key
```

```python
data_url = 'http://www.omdbapi.com/?apikey=' + api_key + '&'
poster_url = 'http://img.omdbapi.com/?apikey=' + api_key + '&'
```

```python
movie = 'Jumanji'
year = 1995

jumanji_url = data_url + 't=' + movie + '&y=' + str(year)
```

```python
data = requests.get(jumanji_url).json()
```

```python
data
```

```python
jumanji_poster_url = poster_url + 'i=' + data['imdbID']
jumanji_poster = requests.get(jumanji_poster_url)
```

```python
# show the image
from IPython.display import Image
Image(jumanji_poster.content)
```
